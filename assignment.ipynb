Question:
Create a 3D array of shape (3, 4, 5) filled with random integers between 1 and 100.
Perform the following tasks:

Replace all elements greater than the mean of the entire array with 0.

Calculate the sum along axis=1 (i.e., sum across the 5 columns in each depth slice).

Extract a new 2D array that contains only the first and last columns from each row of each depth slice.

```python
import numpy as np
arr3d = np.random.randint(1,101,size = (3,4,5))
print(f"3D array:\n {arr3d}")
```

```python
m = np.mean(arr3d)
print(f"Mean:{m}")
u = np.where(arr3d>m,0,arr3d)
print(f"replacing element greater than mean with 0:\n{u}")
```

```python
x=arr3d.sum(axis=1)
print(f"sum of each columns:\n{x}")
```

```python
# y = arr3d[:,:,[0,-1]]
# print(y)
a2d = arr3d.reshape(12,5)
print(f"Converting into 2d array which contains first & last column of 3d array:\n{a2d[:,[0,-1]]}")
```
