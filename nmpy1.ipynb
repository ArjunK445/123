```python
nmp = np.arange(1,51)
arr = nmp.reshape(5,2,5)
print(arr)
print(f"index of 3d :{arr[1,0,3]}")
print(f"slice of 3d:{arr[2:,0:,3:]}")
print()
arr = nmp.reshape(5,10)
print(f"array of 2D:{arr}")
print(f"Indexing an array of 2D:{arr[3,5]}")
print(f"slicing an array of 2D:{arr[0:3,2:7]}")
```
